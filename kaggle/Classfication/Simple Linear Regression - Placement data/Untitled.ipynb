```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#读数据
data_path = r"C:\Users\Leaper\Desktop\ML_CODE\Simple Linear Regression - Placement data\archive (2)\placement.csv"
data = pd.read_csv(data_path)
data.head()
```

```python
#分析数据
#1、查看数据的类型
data.describe()
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
y.shape
```

```python
data['placed'].unique()
# 根据 'placed' 列的值将数据分为两个子集
```

```python

# 根据 'placed' 列的值将数据分为两个子集
placed_data = data[data['placed'] == 1]
not_placed_data = data[data['placed'] == 0]

# 绘制散点图
plt.figure(figsize=(9, 8))
plt.scatter(placed_data["cgpa"], placed_data["placement_exam_marks"], label='Placed', c='blue')
plt.scatter(not_placed_data["cgpa"], not_placed_data["placement_exam_marks"], label='Not Placed', c='red')
# 添加图例和标签
plt.legend()
plt.xlabel('CGPA')
plt.ylabel('Placement Exam Marks')

# 显示图表
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
#model = LinearRegression()
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
accuracy
```

```python
y_pred = model.predict(X_test)
y_pred
```
